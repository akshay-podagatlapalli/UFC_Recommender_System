```python
import pandas as pd
```

```python
data = pd.read_excel("data_i/i_ufc_fight_data.xlsx", sheet_name="fight_data")
```

```python
data.head()
```

```python
# Splitting the 'att' columns
ufc_df = ['R_SIG_STR.', 'B_SIG_STR.', 'R_TOTAL_STR.', 'B_TOTAL_STR.', 'R_TD', 'B_TD', 'R_HEAD', 'B_HEAD', 'R_BODY', 'B_BODY', 'R_LEG', 'B_LEG', 'R_DISTANCE', 'B_DISTANCE', 'R_CLINCH', 'B_CLINCH', 'R_GROUND', 'B_GROUND']
att_columns = ['R_SIG_STR.', 'B_SIG_STR.', 'R_TOTAL_STR.', 'B_TOTAL_STR.', 'R_TD', 'B_TD', 'R_HEAD', 'B_HEAD', 'R_BODY', 'B_BODY', 'R_LEG', 'B_LEG', 'R_DISTANCE', 'B_DISTANCE', 'R_CLINCH', 'B_CLINCH', 'R_GROUND', 'B_GROUND']
for column in att_columns:
    data[column + '_att'] = data[column].str.split(' of ').str[0]
    data[column + '_landed'] = data[column].str.split(' of ').str[1]

# Remove the original columns
data = data.drop(att_columns, axis=1)

data.head(10)
```
